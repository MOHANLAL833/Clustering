```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data=pd.read_csv("Mall_Customers.csv")
data
```

```python
x=data.iloc[:,[3,4]].values
x
```

```python
from sklearn.cluster import SpectralClustering
spc=SpectralClustering(n_clusters=4,random_state=10,assign_labels='discretize').fit(x)
y_cluster=spc.fit_predict(x)
y_cluster
```

```python
supervised=pd.DataFrame(data)
supervised["cluster_group"]=y_cluster
supervised
```

```python
supervised.to_csv("spectral.csv")
```

```python
import seaborn as sns
face=sns.lmplot(data=supervised,x=supervised.columns[3],y=supervised.columns[4],hue=supervised.columns[5],legend=True,
                legend_out=True,fit_reg=False)
```
